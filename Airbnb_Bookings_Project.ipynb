```python
# Which is the most listed room type?
airbnb_roomtype_frequency = dict(airbnb_df.room_type.value_counts())
```

```python
## @title pie_chart visualization for roomtype frequency
# Creating Dataset

room_type = list(airbnb_roomtype_frequency.keys())
data = list(airbnb_roomtype_frequency.values())

# Creating color parameters

colors = ( "#00FF66", "#99FF66", "#FF9966",)

# Creating explode data

explode = (0.03, 0.03, 0.1)

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d})".format(pct, absolute)

# Creating Pie Chart

# Creating plot
fig, airbnb_pie_chart = plt.subplots(figsize =(12, 8))
wedges, texts, autotexts = airbnb_pie_chart.pie(data, autopct = lambda pct: func(pct, data),
                                                explode = explode,
                                                shadow = True,
                                                colors = colors,
                                                startangle = 0,

                                                textprops = dict(color ="black"))

# Adding legend
airbnb_pie_chart.legend(wedges, room_type,
                        title ="Room Type",
                        loc ="upper left",
                        bbox_to_anchor=(1, 0., 0.,1))

plt.setp(autotexts, size = 15, weight = "bold")
airbnb_pie_chart.set_title("Count of Listed Rooms")


```

# Visualizing all listing’s room type proportions from each neighbourhood group using Plotly.Express

```python
# @title Room Type Distribution by Neighbourhood Group
import plotly.express as px

# Define color dictionary
color_dict = {
    "Entire home/apt": "red",
    "Private room": "yellow",
    "Shared room": "green",
    # Add more colors if you have additional room types
}

# Group and calculate room type proportions
room_type_props = (
    airbnb_df.groupby(["neighbourhood_group", "room_type"])
    .size()
    .to_frame(name="Count")
    .reset_index()
)

# Create sunburst chart
fig = px.sunburst(
    room_type_props,
    path=["neighbourhood_group", "room_type"],
    values="Count",
    color="room_type",
    color_discrete_map=color_dict,
    title="Room Type Distribution by Neighbourhood Group",
)

fig.update_layout(margin=dict(t=20, l=12))  # Adjust margins as needed

fig.show()
```

# Finding Top 10 most used words from listing names

```python
#Creting empty list to store name strings
airbnb_names=[]

#Getting name string from 'name' column and appending it to the empty list
for name in airbnb_df.name:
    airbnb_names.append(name)

#Setting a function to split name strings into seperate words
def split_name(name):
    ns = str(name).split()
    return ns

#Creating empty list to store the count of words
names_count = []

#Getting name string to append it to the names_count list
for n in airbnb_names:
    for word in split_name(n):
        word = word.lower()
        names_count.append(word)


# Importing 'counter' library to count and generate raw data and count top 10 most used words
from collections import Counter

# Counting most common words
count_words = Counter(names_count).most_common()
count_words[:10]
```

```python
# Cleaning the List by removing prepositions
items_to_remove = {('in', 16733), ('the', 3869), ('to', 3827), ('of', 2993), ('-', 2272), ('a', 1909)}
top_10_cleaned = [e for e in count_words if e not in items_to_remove]
top_10 = top_10_cleaned[:10]


#converting the data into DataFrame
word_count_df = pd.DataFrame(top_10)
word_count_df.rename(columns={0:'Words',1:'Counts'},inplace=True)
word_count_df
```

## Visualizing the findings

```python
import seaborn as sns
import pandas as pd  # Assuming 'word_count_df' is a pandas DataFrame

# Set figure size and background style
sns.set(rc={'figure.figsize': (12, 6)}, style='white')  # White background

color_palette = sns.color_palette('hls')

# Create the bar chart with bold elements
count_viz = sns.barplot(
    x='Words',
    y='Counts',
    data=word_count_df,
    #palette='Blues',
    palette=color_palette) # Optional color palette (adjust as desired)



# Customize the chart title, axis labels, and bar labels
count_viz.set_title('Top 10 Most Used Words in Listing Names', fontsize=15, weight='bold')
count_viz.set_xlabel('Words', weight='bold')
count_viz.set_ylabel('Count of Words', weight='bold')
count_viz.set_xticklabels(count_viz.get_xticklabels(), rotation=90, weight='bold', size=15)

# Optional: Tighten layout to avoid clipping elements
plt.tight_layout()

plt.show()

```

From the chart above, we see the top 10 words used in the listing name. We can use the word cloud visualization method to help us better understand the chart.

```python
#Creating DataFrame
count_host_id = list(airbnb_df['host_id'].value_counts())
host_id = list(airbnb_df.host_id)
listing_count= list(zip(host_id,count_host_id))
count_host_id_df= pd.DataFrame(listing_count)
count_host_id_df.rename(columns={0:'Host_Id',1:'Counts'},inplace=True)

#Storing top 10 hosts with most listings
top_host_id = count_host_id_df.head(10)
top_host_id
```

```python
import seaborn as sns
import pandas as pd  # Assuming 'top_host_id' is a pandas DataFrame

# Set the desired figure size
sns.set(rc={'figure.figsize': (10, 8)})

# Set the background color as white
sns.set_style('white')

# Generate a random color list for each data point (adjust as needed)
num_bars = len(top_host_id)
color_palette = sns.color_palette('hls', num_bars)  # Adjust palette as desired

# Plot the bar chart with different colors
viz_bar = sns.barplot(
    x='Host_Id',
    y='Counts',
    data=top_host_id,
    order=top_host_id.sort_values('Counts', ascending=False).Host_Id,
    palette=color_palette
)

# Set font size for the title
sns.set(font_scale=2)

# Set chart title, axis labels, and rotate x-axis labels
viz_bar.set_title('Hosts with Most Listings ')
viz_bar.set_xlabel('Host IDs')
viz_bar.set_ylabel('Count of Listings')
viz_bar.set_xticklabels(viz_bar.get_xticklabels(), rotation=30)

plt.tight_layout()  # Adjust spacing to avoid clipping elements

plt.show()
```

##Average Number of Nights Spend Per Location

```python
## @title  Total Number of Nights Spend Per Location
# Calculate total nights spent per neighbourhood group efficiently
total_nights = airbnb_df.groupby('neighbourhood_group')['minimum_nights'].mean()

# Find the neighbourhood with the minimum total nights spent
min_nights_neighbourhood = total_nights.idxmin()
min_nights_value = total_nights.min()

# Create the visualization with a clear variable name
min_nights_by_group = sns.barplot(
    x=total_nights.index,  # Use group index directly for x-axis labels
    y=total_nights.values,  # Use values directly for y-axis data
    data=total_nights.to_frame('minimum_nights').reset_index(),  # Convert to DataFrame for plotting
    color='yellow'  # Adjust color as desired
)

# Customize the plot
min_nights_by_group.set_title('Minimum Total Nights Spent Per Neighbourhood Group')
min_nights_by_group.set_xlabel('Neighbourhood Group')
min_nights_by_group.set_ylabel('Total Minimum Nights')
min_nights_by_group.set_xticklabels(min_nights_by_group.get_xticklabels(), rotation=45, size='15')
plt.tight_layout()
plt.show()


print(f"Minimum Total Nights Spent: {round(min_nights_value,1)}")
```
